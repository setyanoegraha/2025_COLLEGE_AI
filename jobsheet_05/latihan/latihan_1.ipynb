## Import Library

```python
import pandas as pd
```

## Load Dataset

```python
file_path = r'D:\Kuliah\Semester2\Artificial_Intelligence\semua_kode\jobsheet_05\latihan\Data_csv.csv'
df = pd.read_csv(file_path)

df
```

## Mengatasi Missing Values dengan Drop Data

### 1. Hapus Data yang Kosong (NaN)

```python
df = pd.read_csv(file_path)
df.dropna(inplace=True)

df
```

## Mengganti Missing Values dengan Nilai Baru

### 2. Ganti Data yang kosong (NaN) menjadi:
<ul><li>Umur = 50 </li><li>Gaji = 25000000</li></ul>

```python
df = pd.read_csv(file_path)
df.fillna({'Umur': 50, 'Gaji': 25000000}, inplace=True)

df
```

## Mengganti Missing Values dengan Nilai Mean

### 3. Mengganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Rata-Rata

```python
df = pd.read_csv(file_path)
mean_umur = df['Umur'].mean();
mean_gaji = df['Gaji'].mean();

df.fillna({'Umur': mean_umur, 'Gaji': mean_gaji}, inplace=True)

df
```

## Mengatasi Missing dengan Nilai Median

### 4. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Median

```python
df = pd.read_csv(file_path)
median_umur = df['Umur'].median()
median_gaji = df['Gaji'].median()

df.fillna({'Umur': median_umur, 'Gaji': median_gaji}, inplace=True)

df
```

## Mengganti Missing Values dengan Nilai Modus

### 5. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Modus

```python
df = pd.read_csv(file_path)
modus_umur = df['Umur'].mode()[0]
modus_gaji = df['Gaji'].mode()[0]

df.fillna({'Umur': modus_umur, 'Gaji': modus_gaji}, inplace=True)

df
```

## Mengatasi Wrong Data dengan Ubah Data

### 6. Apabila Ada Umur yang diatas 120 maka ubah menjadi 33

```python
df = pd.read_csv(file_path)

for x in df.index:
    if df.loc[x, 'Umur'] > 120:
        df.loc[x, 'Umur'] = 33

df
```
