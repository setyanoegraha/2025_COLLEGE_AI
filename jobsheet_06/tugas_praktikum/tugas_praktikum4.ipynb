# Latihan 4: Penggunaan CPU

Kita memiliki penggunaan CPU rata-rata per jam untuk komputer pekerja selama seminggu. Setiap baris data mewakili satu hari dalam seminggu yang dimulai dengan Senin. Setiap kolom data adalah satu jam dalam sehari dimulai dengan 0 menjadi tengah malam.

Buat bagan yang menunjukkan penggunaan CPU selama seminggu. 

Anda harus dapat menjawab pertanyaan-pertanyaan berikut menggunakan bagan:
>- **Jam berapa pekerja biasanya makan siang?**
>- **Apakah pekerja tersebut bekerja pada akhir pekan?**
>- **Pada hari apa pekerja mulai bekerja pada komputer mereka pada malam hari?**


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data penggunaan CPU
cpu_usage = [
    [2, 2, 4, 2, 4, 1, 1, 4, 4, 12, 22, 23, 45, 9,
        33, 56, 23, 40, 21, 6, 6, 2, 2, 3],  # Monday
    [1, 2, 3, 2, 3, 2, 3, 2, 7, 22, 45, 44, 33, 9,
        23, 19, 33, 56, 12, 2, 3, 1, 2, 2],  # Tuesday
    [2, 3, 1, 2, 4, 4, 2, 2, 1, 2,  5, 31, 54, 7, 6,
        34, 68, 34, 49, 6, 6, 2, 2, 3],  # Wednesday
    [1, 2, 3, 2, 4, 1, 2, 4, 1, 17, 24, 18, 41, 3, 44,
        42, 12, 36, 41, 2, 2, 4, 2, 4],  # Thursday
    [4, 1, 2, 2, 3, 2, 5, 1, 2, 12, 33, 27, 43, 8,
        38, 53, 29, 45, 39, 3, 1, 1, 3, 4],  # Friday
    [2, 3, 1, 2, 2, 5, 2, 8, 4, 2, 3, 1, 5, 1, 2, 3,
        2, 6, 1, 2, 2, 1, 4, 3],         # Saturday
    [1, 2, 3, 1, 1, 3, 4, 2, 3, 1, 2, 2, 5, 3, 2, 1,
        4, 2, 45, 26, 33, 2, 2, 1],       # Sunday
]

# Label
days = ['Monday', 'Tuesday', 'Wednesday',
        'Thursday', 'Friday', 'Saturday', 'Sunday']
hours = list(range(24))

# Plot heatmap
plt.figure(figsize=(16, 6))
sns.heatmap(cpu_usage, cmap='YlOrRd', xticklabels=hours, yticklabels=days,
            linewidths=0.3, linecolor='gray', annot=True, fmt='d')
plt.title('Penggunaan CPU per Jam Selama Seminggu', fontsize=16)
plt.xlabel('Jam (0 = Tengah Malam)', fontsize=12)
plt.ylabel('Hari', fontsize=12)
plt.tight_layout()
plt.show()
```
