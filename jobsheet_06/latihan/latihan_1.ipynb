# 1. Melakukan Visualisasi Data menggunakan Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)
```

```python
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
```

```python
explode = (0, 0, 0, 0.1, 0)
```

```python
plt.pie(votes,
        labels=flavors,
        autopct='%1.2f%%',
        colors=colors,
        explode=explode,
        shadow=True)
plt.show()
```
